```python
import requests

while True:
    prompt = input("You: ")

    if prompt.lower() in ["exit", "quit"]:
        print("Chat ended.")
        break

    response = requests.post(
        "http://localhost:11434/api/generate",
        json={
            "model": "qwen3:8b",
            "prompt": prompt,
            "stream": False
        }
    )

    print("AI:", response.json()["response"])
```
